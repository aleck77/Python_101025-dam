### 1. Создание списка с пользовательскими данными

Напишите программу, которая принимает список целых чисел от пользователя и добавляет каждое число в динамический массив. Программа должна принимать новое число от пользователя и добавлять его в динамический массив до тех пор, пока пользователь не введет команду "Выход". Выведите итоговый динамический массив на экран с помощью команды print.

Пример вывода:

Введите число или 'выход':: 1

Введите число или 'выход':: 2

Введите число или 'выход':: выход

Итоговый динамический массив: [1, 2]

```python
numbers = []
while True:
    user_input = input("Введите число или 'выход':: ")
    if user_input == "выход":
        break
    number = int(user_input)
    numbers += [number]
print("Итоговый динамический массив:", numbers)
```
