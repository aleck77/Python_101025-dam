# 1. Звёздочки вместо чисел
<p>Напишите программу, которая заменяет все цифры в строке на звёздочки *.
<p>text = "My number is 123-456-789"
<p>Пример вывода:

- Строка: My number is 123-456-789
- Результат: My number is ***-***-***

```python
string = "My number is 123-456-789"
print("Строка:", string)
result = string
for char in string:
    if char.isdigit():
        result = result.replace(char, "*")
print("Результат:", result)
```

# 2.Увеличение чисел

Напишите программу, которая делает все целые числа в строке в 10 раз больше.

text = "I have 5 apples and 10 oranges, price is 0.5 each. Card number is ....3672."

Пример вывода:

I have 50 apples and 100 oranges, price is 5.0 each. Card number is ....3672.

```python
text = "I have 5 apples and 10 oranges, price is 0.5 each. Card number is ....3672."
words = text.split()
result_words = []
for word in words:
    if word. isdecimal(): 
        new_number = str(int(word) * 10)
        result_words += [new_number] 
    else:       
        result_words += [word] 
new_text = " ".join(result_words)
print(new_text)
```
