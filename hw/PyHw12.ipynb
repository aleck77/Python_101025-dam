# 1. Сумма положительных чисел

Напишите программу, которая обрабатывает список float чисел, вычисляет сумму положительных чисел.
Результат необходимо вывести в формате денежной суммы (до 2 символов после запятой), добавляя символ валюты "$" вначале
и разделяя тысячи запятой.

- Данные:
<p>numbers = [1245.4435, -302.1403, 87459.99, -520.8265, 1450.001]

- Пример вывода:
<p>Сумма положительных чисел: $90,155.43

```python
numbers = [1245.4435, -302.1403, 87459.99, -520.8265, 1450.001]
positive_sum = 0
for num in numbers:
    if num > 0:
        positive_sum += num
formatted_sum = "${:,.2f}".format(positive_sum)
print(f"Сумма положительных чисел: {formatted_sum}")
```

# 2. Счета клиентов

Напишите программу, которая принимает список строк в формате "name age balance" и выводит информацию о каждом человеке с отформатированными данными: имя — 10 символов, возраст — 3 символа, баланс — 10 символов с двумя знаками после запятой. 
Используйте заранее подготовленный список строк.

Данные:

data_list = [
 "John 23 12345.678",
 "Alice 30 200.50",
 "Bob 35 15000.3",
 "Charlie 40 500.75"
]

Пример вывода:

<p>Имя: John | Возраст: 23 | Баланс: 12345.68
<p>Имя: Alice | Возраст: 30 | Баланс: 200.50
<p>Имя: Bob | Возраст: 35 | Баланс: 15000.30
<p>Имя: Charlie | Возраст: 40 | Баланс: 500.75



```python
data_list = [
    "John 23 12345.678",
    "Alice 30 200.50",
    "Bob 35 15000.3",
    "Charlie 40 500.75"
]

for item in data_list:
    name, age, balance = item.split()
    format_name = f"{name:<10}"
    format_age = f"{int(age):>3}"
    format_balance = f"{float(balance):>10.2f}"
    print(f"Имя: {format_name} | Возраст: {format_age} | Баланс: {format_balance}")
```
