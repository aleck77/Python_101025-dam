#### 1. Проверка на подмножество

Напишите программу, которая проверяет, содержит ли первое множество все элементы второго множества. Реализуйте решение несколькими способами. Решите одним из способов не используя возможности множеств.

Данные:

`set1 = {1, 2, 3, 4}`

`set2 = {2, 3}`


Пример вывода:

True

```python
set1 = {1, 2, 3, 4}
set2 = {2, 3}
print(set1 >= set2)
print(set1.issuperset(set2))
```

```python
set1 = {1, 2, 3, 4}
set2 = {2, 3}
flag = True
for i in set2:
    if i not in set1:
        flag = False
        break
print(flag)

```

```python
set1 = {1, 2, 3, 4}
set2 = {2, 3}
for i in set2:
    if i not in set1:
        print(False)
        break
else:
    print(True)
```

```python
set1 = {1, 2, 3, 4}
set2 = {2, 3}
missing = [x for x in set2 if x not in set1]
print(not bool(missing))
```

#### 2. Зеркальное подмножество

Напишите программу, которая проверяет, являются ли элементы одного из множеств подмножеством другого. В случае положительного ответа возвращает разницу между двумя множествами. Проверить необходимо в обе стороны.

Данные:

`set1 = {2, 3, 4, 5, 6}`

`set2 = {4, 5}`


Пример вывода:

Подмножество: True

Разница: {2, 3, 6}

```python
set1 = {2, 3, 4, 5, 6}
set2 = {4, 5}
if set1 >= set2:
    print('Подмножество:', set1 >= set2)
    print('Разница:', set1 - set2)
elif set1 <= set2:
    print('Подмножество:', set1 <= set2)
    print('Разница:', set2 - set1)
```

```python
set1 = {2, 3, 4, 5, 6}
set2 = {4, 5}
if set1.issuperset(set2):
    print('Подмножество:', set1.issuperset(set2))
    print('Разница:', set1 ^ set2)
elif set2.issuperset(set1):
    print('Подмножество:', set2.issuperset(set1))
    print('Разница:', set2.symmetric_difference(set1))
```
