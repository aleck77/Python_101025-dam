#### 1. Реверс словаря

Напишите программу, которая меняет местами ключи и значения в словаре. Если значения повторяются, добавьте их ключи в список.

Данные:

`data = {"a": 1, "b": 2, "c": 1, "d": 3}`

* Пример вывода:

Перевернутый словарь: {1: ['a', 'c'], 2: ['b'], 3: ['d']}

#### 1. Способ

```python
data = {"a": 1, "b": 2, "c": 1, "d": 3}
result = {}

for key, value in data.items():

    result.setdefault(value, []).append(key)

print('Перевернутый словарь:', result)
```

#### 2. Способ

```python
data = {"a": 1, "b": 2, "c": 1, "d": 3}
result = {}

for key, value in data.items():  # [2]
    if value in result:  # Проверяем, есть ли уже такой ключ в новом словаре [1]
        result[value].append(key)  # Если есть, добавляем старый ключ в список [3]
    else:
        result[value] = [key]  # Если нет, создаем новый список с этим ключом [4]
print('Перевернутый словарь:', result)
```

#### 2. Распределение студентов по группам
У вас есть словарь, где ключи — имена студентов, а значения — их баллы за экзамен.
Необходимо распределить студентов по группам:

* "Отличники": баллы >= 85
* "Хорошисты": баллы от 70 до 84
* "Троечники": баллы от 50 до 69
* "Не сдали": баллы < 50

Создайте словарь с ключами-группами и количеством студентов в группе в качестве значений.

Данные:

`students = {"Аня": 92, "Боря": 76, "Ваня": 65, "Галя": 48, "Дима": 88, "Ева": 54}`

`groups = ["Отличники", "Хорошисты", "Троечники", "Не сдали"]`


Пример вывода:

Распределение по группам:

{'Отличники': 2, 'Хорошисты': 1, 'Троечники': 2, 'Не сдали': 1}

#### 1. Способ

```python
students = {"Аня": 92, "Боря": 76, "Ваня": 65, "Галя": 48, "Дима": 88, "Ева": 54}
groups = ["Отличники", "Хорошисты", "Троечники", "Не сдали"]

print("Распределение по группам:")
key_grope_cnt = {}
for n in groups:
    key_grope_cnt[n] = key_grope_cnt.setdefault(n, 0)
for value in students.values():
    if value >= 85:
        key_grope_cnt["Отличники"] += 1
    elif value >= 70:
        key_grope_cnt["Хорошисты"] += 1
    elif value >= 50:
        key_grope_cnt["Троечники"] += 1
    else:
        key_grope_cnt["Не сдали"] += 1
print(key_grope_cnt)
```

#### 2. Способ

```python
students = {"Аня": 92, "Боря": 76, "Ваня": 65, "Галя": 48, "Дима": 88, "Ева": 54}
groups = ["Отличники", "Хорошисты", "Троечники", "Не сдали"]
print("Распределение по группам:")
key_grope_cnt = dict.fromkeys(groups, 0)
for value in students.values():
    if value >= 85:
        key_grope_cnt["Отличники"] += 1
    elif value >= 70:
        key_grope_cnt["Хорошисты"] += 1
    elif value >= 50:
        key_grope_cnt["Троечники"] += 1
    else:
        key_grope_cnt["Не сдали"] += 1
print(key_grope_cnt)
```
