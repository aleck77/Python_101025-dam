#### 1. Повторения букв
Реализуйте функцию, которая принимает текст и возвращает словарь с подсчётом количества каждой буквы, игнорируя регистр.
Данные:

`text = "Programming is fun!"`

Пример вывода:

`{'p': 1, 'r': 2, 'o': 1, 'g': 2, 'a': 1, 'm': 2, 'i': 2, 'n': 2, 's': 1, 'f': 1, 'u': 1}`

```python
from collections import Counter
import time

def cnt_alpha(text):
    words = text.lower().replace(" ", "").replace("!", "")
    word_count = Counter(words)
    return word_count

text = "Programming is fun!"

start_time = time.time()
result = cnt_alpha(text)
end_time = time.time()
print(f"Время рвботы ф-ции: {end_time - start_time:.10f} секунд")

print(dict(result))
```

```python
import time
from collections import Counter

text = "Programming is fun!"

start_time = time.time()
word_count = Counter([char.lower() for char in text if char.isalpha()])
end_time = time.time()
print(f"Время рвботы с list comprehension: {end_time - start_time:.10f} секунд")

print(dict(word_count))
```

```python
import time
from collections import Counter

text = "Programming is fun!"

start_time = time.time()
words = text.lower().replace(" ", "").replace("!", "")
word_count = Counter(words)
end_time = time.time()
print(f"Время рвботы без list comprehension: {end_time - start_time:.10f} секунд")

print(dict(word_count))
```

#### 2. Группировка студентов по классам

Создайте структуру для группировки студентов по классам.
Добавьте студентов в соответствующие группы.

Данные:

`students = [("class1", "Alice"), ("class2", "Bob"), ("class1", "Charlie"), ("class3", "Daisy")]`


Пример вывода:

`{'class1': ['Alice', 'Charlie'], 'class2': ['Bob'], 'class3': ['Daisy']}`

```python
from collections import defaultdict

def group_students_by_claster(students):
    claster_dict = defaultdict(list)
    for claster, name in students:
        claster_dict[claster].append(name)
    return dict(claster_dict)

students = [("class1", "Alice"), ("class2", "Bob"), ("class1", "Charlie"), ("class3", "Daisy")]

result = group_students_by_claster(students)

print(result)
```
