2. Сумма вложенных чисел
Напишите функцию, которая принимает список словарей, где каждый словарь содержит имя пользователя и список баллов. Функция должна вернуть сумму всех чисел. Добавьте документацию и аннотации типов для всех параметров и возвращаемого значения.

Данные:

data = [

    {"name": "Alice", "scores": [10, 20, 30]},

    {"name": "Bob", "scores": [5, 15, 25]},

    {"name": "Charlie", "scores": [7, 17, 27]}

]

Пример вывода:

Итоговый балл: 156

1 вариант

```python
from typing import List, Dict

def total_sum(lst: List[Dict[str, List[int]]]) -> int:
    """
    Принимает список словарей с именами и баллами, 
    возвращает общую сумму всех баллов.
    """
    return sum(sum(dct['scores']) for dct in lst)

data = [
    {"name": "Alice", "scores": [10, 20, 30]},
    {"name": "Bob", "scores": [5, 15, 25]},
    {"name": "Charlie", "scores": [7, 17, 27]}
]

print(f"Итоговый балл: {total_sum(data)}")
```

2 вариант

```python
from typing import List, Dict

def total_sum(data: List[Dict[str, List[int]]]) -> int:
    """
    Принимает список словарей с именами и баллами, 
    возвращает общую сумму всех баллов.
    """
    return sum(
        score
            for student in data
                for score in student["scores"]
    )

data = [
    {"name": "Alice", "scores": [10, 20, 30]},
    {"name": "Bob", "scores": [5, 15, 25]},
    {"name": "Charlie", "scores": [7, 17, 27]}
]
print(f"Итоговый балл: {total_sum(data)}")
```
