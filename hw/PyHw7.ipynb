### 1. Сумма цифр числа
Напишите программу, которая вычисляет сумму цифр введённого числа.
Подсказка: последнюю цифру числа можно найти с помощью остатка от деления на 10 (num % 10), а удалить ее из числа можно с помощью деления на 10 нацело (num // 10).

Пример вывода:
* Введите число: 12345
* Сумма цифр: 15

```python
num = int(input('Введите число:'))
s = 0
while num > 0:
    d = num % 10
    num = num // 10
    s += d
print('Сумма цифр:', s)
```

### 2. Палиндром

Напишите программу, которая запрашивает у пользователя целое число и определяет, является ли оно палиндромом. Число является палиндромом, если оно читается одинаково слева направо и справа налево.

Пример вывода:

* Введите число: 12321
* Число 12321 является палиндромом.

* Введите число: 1234
* Число 1234 не является палиндромом.

```python
user_num = abs(int(input('Введите число:')))
num = user_num
revers = 0
while num > 0:
    d = num % 10
    num = num // 10
    revers = revers * 10 + d
if user_num == revers:
    print('Число ', user_num, 'является палиндромом.')
else:
    print('Число ', user_num, 'не является палиндромом.')
```

print(rnd_num, count)

```python
import random

count = 0
rnd_num = random.randint(1, 100)
while count < 10:
    count += 1
    print(rnd_num, count)
    user_num = abs(int(input('Угадайте число от 1 до 100. У вас 10 попыток.')))
    if user_num == rnd_num and count == 1:
        print('Ваше предположение:', user_num, '\n Поздравляем! Вы угадали число:', rnd_num, 'за', count , 'попыток. Отличный результат!\n Отличная интуиция! Вы угадали с первой попытки!')
        break
    elif user_num == rnd_num and count > 1:
        print('Ваше предположение:', user_num, '\n Поздравляем! Вы угадали число:', rnd_num, '.\n Вы угадали число за', count , 'попытки. Отличный результат!')
        print('Вы использовали', count,'попыток.\n Загаданное число -', rnd_num)
        break
    elif user_num < rnd_num:
        print('Ваше предположение:', user_num,'\n' , 'Загаданное число больше вашего')
    else:
        print('Ваше предположение:', user_num,'\n' , 'Загаданное число меньше вашего')
```
