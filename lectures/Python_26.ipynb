````python
import sys
import json

# –£–ù–ò–í–ï–†–°–ê–õ–¨–ù–ê–Ø –§–£–ù–ö–¶–ò–Ø –¥–ª—è –ø—Ä–æ—Å–º–æ—Ç—Ä–∞ –ª—é–±–æ–≥–æ –º–æ–¥—É–ª—è
def show_module_info(module, show_details=True):
    """
    –£–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –ª—é–±–æ–≥–æ –º–æ–¥—É–ª—è
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    module - –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –º–æ–¥—É–ª—å (–Ω–∞–ø—Ä–∏–º–µ—Ä, os, sys, json)
    show_details - –ø–æ–∫–∞–∑—ã–≤–∞—Ç—å –ª–∏ –¥–µ—Ç–∞–ª—å–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é
    """
    print("=" * 70)
    print(f"–ê–ù–ê–õ–ò–ó –ú–û–î–£–õ–Ø: {module.__name__}")
    print("=" * 70)
    
    # 1. –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É –º–æ–¥—É–ª—è (–µ—Å–ª–∏ –µ—Å—Ç—å)
    file_path = getattr(module, '__file__', None)
    if file_path:
        print(f"üìÅ –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É: {file_path}")
    else:
        print(f"üìÅ –í—Å—Ç—Ä–æ–µ–Ω–Ω—ã–π –º–æ–¥—É–ª—å (–±–µ–∑ —Ñ–∞–π–ª–∞)")
    print()
    
    # 2. –î–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏—è –º–æ–¥—É–ª—è
    if module.__doc__:
        doc_preview = module.__doc__[:200].replace('\n', ' ')
        print(f"üìù –û–ø–∏—Å–∞–Ω–∏–µ: {doc_preview}...")
    print()
    
    # 3. –ü–æ–ª—É—á–∞–µ–º –≤—Å–µ –∞—Ç—Ä–∏–±—É—Ç—ã
    all_attrs = dir(module)
    
    # 4. –§–∏–ª—å—Ç—Ä—É–µ–º —Ç–æ–ª—å–∫–æ –ø—É–±–ª–∏—á–Ω—ã–µ (–±–µ–∑ _ –≤ –Ω–∞—á–∞–ª–µ)
    public_attrs = [name for name in all_attrs if not name.startswith('_')]
    
    # 5. –†–∞–∑–¥–µ–ª—è–µ–º –Ω–∞ —Ñ—É–Ω–∫—Ü–∏–∏ –∏ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—ã
    functions = []
    constants = []
    classes = []
    
    for name in public_attrs:
        obj = getattr(module, name)
        if callable(obj):
            if isinstance(obj, type):
                classes.append(name)
            else:
                functions.append(name)
        else:
            constants.append(name)
    
    # 6. –í—ã–≤–æ–¥–∏–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É
    print(f"üìä –°–¢–ê–¢–ò–°–¢–ò–ö–ê:")
    print(f"   –í—Å–µ–≥–æ –∞—Ç—Ä–∏–±—É—Ç–æ–≤: {len(all_attrs)}")
    print(f"   –ü—É–±–ª–∏—á–Ω—ã—Ö: {len(public_attrs)}")
    print(f"   ‚îú‚îÄ –§—É–Ω–∫—Ü–∏–π: {len(functions)}")
    print(f"   ‚îú‚îÄ –ö–ª–∞—Å—Å–æ–≤: {len(classes)}")
    print(f"   ‚îî‚îÄ –ö–æ–Ω—Å—Ç–∞–Ω—Ç/–ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö: {len(constants)}")
    print()
    
    if show_details:
        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ 20 —Ñ—É–Ω–∫—Ü–∏–π
        print(f"üîß –§–£–ù–ö–¶–ò–ò (–ø–µ—Ä–≤—ã–µ 20 –∏–∑ {len(functions)}):")
        for i, name in enumerate(functions[:20], 1):
            print(f"   {i:2}. {name}")
        
        if len(functions) > 20:
            print(f"   ... –∏ –µ—â—ë {len(functions) - 20} —Ñ—É–Ω–∫—Ü–∏–π")
        print()
        
        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –∫–ª–∞—Å—Å—ã
        if classes:
            print(f"üèóÔ∏è  –ö–õ–ê–°–°–´ ({len(classes)}):")
            for i, name in enumerate(classes, 1):
                print(f"   {i:2}. {name}")
            print()
        
        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ 15 –∫–æ–Ω—Å—Ç–∞–Ω—Ç
        if constants:
            print(f"üìå –ö–û–ù–°–¢–ê–ù–¢–´/–ü–ï–†–ï–ú–ï–ù–ù–´–ï (–ø–µ—Ä–≤—ã–µ 15 –∏–∑ {len(constants)}):")
            for i, name in enumerate(constants[:15], 1):
                value = getattr(module, name)
                value_str = str(value)[:50]
                print(f"   {i:2}. {name} = {value_str}")
            
            if len(constants) > 15:
                print(f"   ... –∏ –µ—â—ë {len(constants) - 15} –∫–æ–Ω—Å—Ç–∞–Ω—Ç")

# –¢–µ—Å—Ç–∏—Ä—É–µ–º –Ω–∞ —Ä–∞–∑–Ω—ã—Ö –º–æ–¥—É–ª—è—Ö
print("\n" + "="*70)
print("–ü–†–ò–ú–ï–† 1: –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–π –º–æ–¥—É–ª—å sys")
print("="*70)
show_module_info(sys, show_details=True)

print("\n\n" + "="*70)
print("–ü–†–ò–ú–ï–† 2: –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–π –º–æ–¥—É–ª—å json")
print("="*70)
show_module_info(os, show_details=True)
````

–§—É–Ω–∫—Ü–∏—è `os.walk` –ø–æ–∑–≤–æ–ª—è–µ—Ç —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ –æ–±–æ–π—Ç–∏ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—é –∏ –µ—ë —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ, –≤–∫–ª—é—á–∞—è —Ñ–∞–π–ª—ã –∏ –ø–æ–¥–¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏. –≠—Ç–æ —É–¥–æ–±–Ω—ã–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å–æ —Å–ª–æ–∂–Ω—ã–º–∏ —Å—Ç—Ä—É–∫—Ç—É—Ä–∞–º–∏ —Ñ–∞–π–ª–æ–≤–æ–π —Å–∏—Å—Ç–µ–º—ã.


````python
# –†–µ–∫—É—Ä—Å–∏–≤–Ω—ã–π –æ–±—Ö–æ–¥ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏
for root, dirs, files in os.walk("."):
    print(f"–¢–µ–∫—É—â–∞—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è: {root}")
    print(f"–ü–æ–¥–¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏: {dirs}")
    print(f"–§–∞–π–ª—ã: {files}")
    print("-" * 50)

````

–ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è: –ü–æ–∏—Å–∫ —Ñ–∞–π–ª–æ–≤ —Å –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã–º —Ä–∞—Å—à–∏—Ä–µ–Ω–∏–µ–º


````python
# –ü–æ–∏—Å–∫ –≤—Å–µ—Ö .txt —Ñ–∞–π–ª–æ–≤
for root, dirs, files in os.walk("."):
    for file in files:
        if file.endswith(".txt"):
            print(f"–ù–∞–π–¥–µ–Ω —Ñ–∞–π–ª: {os.path.join(root, file)}")
````

os.walk() —Ä–∞–±–æ—Ç–∞–µ—Ç —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ, –Ω–æ –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ –æ–Ω –≤—ã–¥–∞–µ—Ç:

–î–æ–ø—É—Å—Ç–∏–º, —É –Ω–∞—Å —Ç–∞–∫–∞—è —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ –ø–∞–ø–æ–∫:

### –ö–∞–∫ –ø–æ–ª—É—á–∏—Ç—å –≤—Å–µ —Ñ–∞–π–ª—ã –∏–∑ –≤—Å–µ—Ö –ø–æ–¥–¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–π?  
–ù—É–∂–Ω–æ —Å–æ–±—Ä–∞—Ç—å –∏—Ö –≤—Ä—É—á–Ω—É—é, –æ–±—ä–µ–¥–∏–Ω—è—è root —Å –∏–º–µ–Ω–∞–º–∏ —Ñ–∞–π–ª–æ–≤:

````python
import os

all_files = []
for root, dirs, files in os.walk('C:\\Users\\dplog\\Python\\101025dam'):
    for file in files:
        full_path = os.path.join(root, file)
        all_files.append(full_path)

print("–í—Å–µ —Ñ–∞–π–ª—ã:")
print(*all_files, sep='\n')
````

# –ú–æ–¥—É–ª—å sys
–ú–æ–¥—É–ª—å sys –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª—è–µ—Ç –¥–æ—Å—Ç—É–ø –∫ —Ñ—É–Ω–∫—Ü–∏—è–º –∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º, –∫–æ—Ç–æ—Ä—ã–µ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –∏ –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞—é—Ç—Å—è –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ç–æ—Ä–æ–º Python. –≠—Ç–æ—Ç –º–æ–¥—É–ª—å –ø–æ–∑–≤–æ–ª—è–µ—Ç –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–æ–≤–∞—Ç—å —Å –æ–∫—Ä—É–∂–µ–Ω–∏–µ–º Python, —É–ø—Ä–∞–≤–ª—è—Ç—å –∞—Ä–≥—É–º–µ–Ω—Ç–∞–º–∏ –∫–æ–º–∞–Ω–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–∏, –ø—É—Ç—è–º–∏ –ø–æ–∏—Å–∫–∞ –º–æ–¥—É–ª–µ–π, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –≤–≤–æ–¥–æ–º –∏ –≤—ã–≤–æ–¥–æ–º.
–î–ª—è —Ä–∞–±–æ—Ç—ã —Å —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –º–æ–¥—É–ª—è sys –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –µ–≥–æ –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å:

### –ê—Ä–≥—É–º–µ–Ω—Ç—ã –∫–æ–º–∞–Ω–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–∏
`sys.argv` ‚Äî —ç—Ç–æ —Å–ø–∏—Å–æ–∫, –∫–æ—Ç–æ—Ä—ã–π —Å–æ–¥–µ—Ä–∂–∏—Ç –∞—Ä–≥—É–º–µ–Ω—Ç—ã, –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã–µ —Å–∫—Ä–∏–ø—Ç—É –ø—Ä–∏ –µ–≥–æ –∑–∞–ø—É—Å–∫–µ –∏–∑ –∫–æ–º–∞–Ω–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–∏.
* –ü–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç —Å–ø–∏—Å–∫–∞ (`sys.argv[0]`) ‚Äî —ç—Ç–æ –∏–º—è —Å–∫—Ä–∏–ø—Ç–∞.
* –û—Å—Ç–∞–ª—å–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã ‚Äî —ç—Ç–æ –∞—Ä–≥—É–º–µ–Ω—Ç—ã, –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã–µ –ø–æ—Å–ª–µ –∏–º–µ–Ω–∏ —Å–∫—Ä–∏–ø—Ç–∞.


````python
import sys


# –í—ã–≤–æ–¥ –≤—Å–µ—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ –∫–æ–º–∞–Ω–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–∏
print("–í—Å–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã:", sys.argv)


# –†–∞–±–æ—Ç–∞ —Å –ø–µ—Ä–≤—ã–º –∞—Ä–≥—É–º–µ–Ω—Ç–æ–º (–µ—Å–ª–∏ –æ–Ω –ø–µ—Ä–µ–¥–∞–Ω)
if len(sys.argv) > 1:
    print(f"–ü–µ—Ä–µ–¥–∞–Ω–Ω—ã–π –∞—Ä–≥—É–º–µ–Ω—Ç: {sys.argv[1]}")
else:
    print("–ê—Ä–≥—É–º–µ–Ω—Ç—ã –Ω–µ –ø–µ—Ä–µ–¥–∞–Ω—ã.")
````

#–ó–∞–ø—É—Å–∫ python —Ñ–∞–π–ª–∞ —á–µ—Ä–µ–∑ —Ç–µ—Ä–º–∏–Ω–∞–ª:
# –ó–∞–º–µ–Ω–∏—Ç–µ script.py –Ω–∞ –Ω–∞–∑–≤–∞–Ω–∏–µ –≤–∞—à–µ–≥–æ —Ñ–∞–π–ª–∞
python script.py argument1


````python
#–ü—Ä–∏–º–µ—Ä: –û–±—Ä–∞–±–æ—Ç–∫–∞ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤
import sys


if len(sys.argv) > 1:
    for i, arg in enumerate(sys.argv[1:], start=1):
        print(f"–ê—Ä–≥—É–º–µ–Ω—Ç {i}: {arg}")
else:
    print("–ê—Ä–≥—É–º–µ–Ω—Ç—ã –Ω–µ –ø–µ—Ä–µ–¥–∞–Ω—ã.")
````

### –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ sys.argv
–ê—Ä–≥—É–º–µ–Ω—Ç—ã –∫–æ–º–∞–Ω–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –¥–ª—è –ø–µ—Ä–µ–¥–∞—á–∏ –¥–∞–Ω–Ω—ã—Ö –≤ —Å–∫—Ä–∏–ø—Ç, –¥–µ–ª–∞—è –ø—Ä–æ–≥—Ä–∞–º–º—É –±–æ–ª–µ–µ –≥–∏–±–∫–æ–π.


–ü—Ä–∏–º–µ—Ä: –í—ã–≤–æ–¥ —Å–æ–¥–µ—Ä–∂–∏–º–æ–≥–æ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏
–°–∫—Ä–∏–ø—Ç –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—é –∫–∞–∫ –∞—Ä–≥—É–º–µ–Ω—Ç –∏ –≤—ã–≤–æ–¥–∏—Ç –µ—ë —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ.


````python
import sys
import os


# –ü—Ä–æ–≤–µ—Ä—è–µ–º, –ø–µ—Ä–µ–¥–∞–Ω –ª–∏ –∞—Ä–≥—É–º–µ–Ω—Ç
if len(sys.argv) != 2:
    print("–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ: python script.py <–¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è>")
    sys.exit()


directory = sys.argv[1]


# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏
if not os.path.isdir(directory):
    print(f"–î–∏—Ä–µ–∫—Ç–æ—Ä–∏—è '{directory}' –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç.")
    sys.exit()


# –í—ã–≤–æ–¥–∏–º —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏
print(f"–°–æ–¥–µ—Ä–∂–∏–º–æ–µ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏ '{directory}':")
for item in os.listdir(directory):
    print(f"- {item}")
````
