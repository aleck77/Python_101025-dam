Звёздочки вместо чисел
Напишите программу, которая заменяет все цифры в строке на звёздочки *.
text = "My number is 123456789"
Пример вывода:
Строка: My number is 123456789
Результат: My number is ***-***-***


```python
text = "My number is 123-456-789"
print("Строка:", text)
result = text
for char in text:
    if char.isdigit():
        result = result.replace(char, "*")
print("Результат:", result)
```

Увеличение чисел
Напишите программу, которая делает все целые числа в строке в 10 раз больше.
text = "I have 5 apples and 10 oranges, price is 0.5 each. Card number is ....3672."

Пример вывода:
I have 50 apples and 100 oranges, price is 5.0 each. Card number is ....3672.

1

```python
text = "I have 5 apples and 10 oranges, price is 0.5 each. Card number is ....3672."
words = text.split()
for i in range(len(words)):
    if words[i].isdecimal():
        words[i] = str(int(words[i]) * 10)
new_text = " ".join(words)
print(new_text)
```

```python
text = "I have 5 apples and 10 oranges, price is 5.0 each. Card number is ....3672.".split()
result = ''

for word in text:
    if word.isdigit():
        word = str(int(word)*10)
    if "." in word:
        parts = word.split(".")
        if len(parts) == 2 and parts[0].isdigit() and parts[1].isdigit():
            word = str(float(word)*10)
    result += word + ' '
print(result)
```

```python
text = "I have 5 apples and 10 oranges, price is 0.5 each. Card number is ....3672."
# 1. Разбиваем строку на список слов по пробелам
words = text.split()
result_words = []

# 2. Перебираем слова
for word in words:
    # Проверяем, состоит ли слово только из целых чисел (цифр), игнорируя числа с плавающей точкой
    if word.isdecimal(): # [2]
        # Преобразуем в int, умножаем на 10 и преобразуем обратно в str [2]
        result_words.append(str(int(word) * 10))
    else:
        # Оставляем остальные слова как есть
        result_words.append(word)

# 3. Объединяем слова обратно в строку, используя пробел в качестве разделителя [3]
new_text = " ".join(result_words)
print(new_text)
```
