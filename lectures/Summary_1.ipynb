6. "Генератор поздравления"

Условие: Создайте программу, которая генерирует поздравительную открытку.

Запросите у пользователя:

Имя именинника

Возраст

Имя отправителя

Выведите открытку, используя рамку из символов (например, * или =). Внутри рамки должен быть текст поздравления, который использует все введенные данные и переносы строк.

Пример работы:

Ввод: Мария, 30, Анна

Вывод:

```
*********************
* С Днем Рождения,  *
* дорогая Мария!    *
*                   *
* Желаю счастья в   *
* твои 30 лет!      *
*                   *
* Твоя Анна         *
*********************   
```

````python
name = input('Имя именинника: ')
age = input('Возраст: ')
name_out = input('Имя отправителя: ')

print('*********************\n* С Днем Рождения,  *\n* дорогая '+name+'!    *\n*                   *\n* Желаю счастья в   *\n* твои '+age+' лет!      *\n*                   *\n* Твоя '+name_out+'         *\n*********************' )
````

````python
def print_card(name, age, name_out):
    """Печатает поздравительную открытку с динамическим форматированием."""
    
    # Формируем строки открытки
    lines = [
        'С Днем Рождения,',
        f'дорогая {name}!',
        '',
        'Желаю счастья в',
        f'твои {age} лет!',
        '',
        f'Твоя {name_out}'
    ]
    
    # Находим максимальную ширину + отступы
    max_len = max(len(line) for line in lines)
    width = max_len + 4  # 2 пробела слева + 2 справа
    
    # Минимальная ширина для красоты
    width = max(width, 25)
    
    # Печатаем рамку сверху
    print('*' * width)
    
    # Печатаем каждую строку с выравниванием по центру
    for line in lines:
        print(f'* {line.center(width - 4)} *')
    
    # Печатаем рамку снизу
    print('*' * width)

# Получаем данные
name = input('Имя именинника: ')
age = input('Возраст: ')
name_out = input('Имя отправителя: ')

# Выводим открытку
print_card(name, age, name_out)
````
