```python
import os
import sys

# Показываем путь к модулю os
print("Путь к модулю os:")
print(os.__file__)
print()

# Показываем все пути, где Python ищет модули
print("Список путей sys.path:")
for i, path in enumerate(sys.path):
    print(f"{i}: {path}")
```

```python
import os

# Показываем все функции и классы модуля os
print("Все функции и классы модуля os:")
print("=" * 50)

# Получаем список всех атрибутов модуля
all_names = dir(os)

# Фильтруем только функции и классы (без приватных и встроенных)
public_names = [name for name in all_names if not name.startswith('_')]

# Выводим постранично для удобства
for i, name in enumerate(public_names, 1):
    print(f"{i:3}. {name}")

print()
print(f"\nВсего публичных функций/атрибутов: {len(public_names)}")
print()

# Показываем документацию модуля
print("Документация модуля os (первая часть):")
print("=" * 50)
print(os.__doc__[:1000] if os.__doc__ else "Нет документации")
```
