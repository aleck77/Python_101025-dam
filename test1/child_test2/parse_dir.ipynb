```python
import os

def is_file_or_dir(path):
    if os.path.isfile(path):
        print(f"'{path}' существует и это файл.")
    elif os.path.isdir(path):
        print(f"'{path}' существует и это директория.")

dir_path = os.getcwd()
contents = os.listdir(dir_path)

for item in contents:
    full_path = os.path.join(dir_path, item)
    is_file_or_dir(full_path)
```

```python
import os
import sys

if len(sys.argv) < 2:
    print("Ошибка: укажите путь к директории.")
    sys.exit(1)

path = sys.argv[1]  # Исправлено: [1] вместо [4]

if os.path.exists(path) and os.path.isdir(path):
    print(f"Содержимое директории '{path}':\n")  # Добавлена пустая строка
    
    items = os.listdir(path)
    
    folders = [i for i in items if os.path.isdir(os.path.join(path, i))]
    files = [i for i in items if os.path.isfile(os.path.join(path, i))]
    
    # Добавлена сортировка
    folders.sort()
    files.sort()

    print("Папки:")
    for f in folders:
        print(f"- {f}")

    print("\nФайлы:")  # Добавлена пустая строка перед файлами
    for f in files:
        print(f"- {f}")
else:
    print(f"Ошибка: путь '{path}' не найден или не является директорией.")
```
