### Задание 4.
Напишите программу, которая запрашивает расстояние (в километрах) и среднюю скорость автомобиля (в км/ч). 

Программа должна вывести время, которое потребуется, чтобы преодолеть указанное расстояние.
### Пример вывода:
- Введите расстояние (в км): 150
- Введите среднюю скорость (в км/ч): 50
- Время в пути: 3.0 часов

```python
distance = int(input('Введите расстояние (в км):'))
speed = int(input('Введите среднюю скорость (в км/ч):'))
print('Время в пути:', distance / speed, ' часов', sep='')
```

### Задание 4.1*
Доработайте программу, чтобы она выводила время, которое потребуется, чтобы преодолеть указанное расстояние в часах и минутах (без дробной части).
### Пример вывода:
- Введите расстояние (в км): 400
- Введите среднюю скорость (в км/ч): 123
- Время в пути: 3 часа и 15 минут

```python
distance = int(input('Введите расстояние (в км):'))
speed = int(input('Введите среднюю скорость (в км/ч):'))
hours = distance // speed
minutes = distance % speed / speed * 60
print('Время в пути:', hours, ' часа и ', int(minutes), ' минут', sep='')
```

### Задание 5.
Напишите программу, которая принимает от пользователя количество дней до события и выводит, сколько это недель и дней.
### Пример вывода:
- Введите количество дней до события: 45
- До события осталось: 6 недель и 3 дня.

```python
number_days = int(input('Введите количество дней до события:'))
weeks = number_days // 7
days = number_days % 7
days = number_days % 7
print('До события осталось:', weeks, ' недель и ', days, ' дня.', sep='')
```

### Задание 6.

### Задание 8.
Напишите программу, которая считывает число "n" и вычисляет выражение "n + nn +
nnn", где "nn" — это число "n", записанное дважды, а "nnn" — трижды.
### Пример вывода:
- Введите число: 5
- Значение выражения: 615

```python
n = input('Введите число:')

nn = n + n
nnn = n + n + n
result = int(n) + int(nn) + int(nnn)
print('Значение выражения:', result)
```

```python
n = input('Введите число:')
nn = n * 11
nnn = n * 111
result = int(n) + int(nn) + int(nnn)
print('Значение выражения:', result)
```

### Задание 9.
Напишите программу, которая считывает четырёхзначное число и меняет местами
первую и последнюю цифры.
### Пример вывода:
- Введите четырёхзначное число: 1234
- Число после изменения: 4231

```python
a = int(input('Enter a four-digit number:'))
dig1 = a // 1000
dig2 = a // 100 % 10
dig3 = a // 10 % 10
dig4 = a % 10
dig1, dig4 = dig4, dig1
print('Sum of the digits of a number: ', dig1, dig2, dig3, dig4, sep='')
```
