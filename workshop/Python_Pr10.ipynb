````python
from collections import Counter

text = "This is a test. This test is only a test."

def counter_alp(text):
    cnt_alp = Counter(text.lower().replace('.','').replace(',','').replace(' ',''))
    return cnt_alp.most_common()

print(dict(counter_alp(text)))
````

````python
from collections import Counter

text = "This is a test. This test is only a test."

def counter_alp(text):
    return Counter([ch for ch in text.lower() if ch.isalpha()]).most_common()
    
print(dict(counter_alp(text)))
````

3. ***Популярные слова***  
Реализуйте функцию, которая принимает любое количество строк с текстом.  
Функция должна возвращать подсчет самых популярных слов в количестве, переданном в функцию. Программа должна игнорировать регистр слов. Выведите 5 самых популярных слов и их количество.  
*Данные:*  
`text1 = "This is a sample text with some repeated words."`  
`text2 = "Another sample text with different words."`  
`text3 = "Text processing is fun when words repeat."`  
*Пример вывода:*  
5 самых популярных слов:  
```
	words: 3
	text: 3
	with: 2
	sample: 2
	is: 2
```

````python

text1 = "This is a sample text with some repeated words."
text2 = "Another sample text with different words."
text3 = "Text processing is fun when words repeat."

def clean_text(text):
    return text.lower().replace('.', '').replace(',', '')

def print_top_words(dct):
    print('5 самых популярных слов:')
    for word, count in dct.items():
        print(f'{word}:{count}')
def count_words(*args):
    t1 = Counter('')
    for text in args:
        t1 += Counter(clean_text(text).split())
        return dict(t1.most_common(5))
    
print_top_words(count_words(text1, text2, text3))
````

Реализуйте функцию, которая принимает словарь задач с категориями и группирует задачи по их категориям.  
*Данные:*  
```
tasks = {
    "task1": "работа",
    "task2": "учёба",
    "task3": "развлечения",
    "task4": "работа",
    "task5": "учёба"
}
```
*Пример вывода:*   
Группировка по категориям:  
```
{
    'работа': ['task1', 'task4'],
    'учёба': ['task2', 'task5'],
    'развлечения': ['task3']
}
```

````python
from collections import defaultdict

dd = defaultdict(list)

tasks = {
    "task1": "работа",
    "task2": "учёба",
    "task3": "развлечения",
    "task4": "работа",
    "task5": "учёба"
}

def grouped_work(tasks):
    for task, work in tasks.items():
        dd[work].append(task)
    return dict(dd)

def print_tasks(dct):
    print('Группировка по категориям:')
    print(dct)

print_tasks(grouped_work(tasks))
````
