Напишите рекурсивную функцию fib(n), которая возвращает n-е число Фибоначчи.

````python
def f(n):
    if n == 1:
        return 0
    if n == 2:
        return 1
    return f(n-1) + f(n-2)

f(5)
````

Напишите сумму чисел от 1 до n рекурсивно

````python
def s(n):
    if n==1:
        return 1
    return n + s(n-1)

print(s(5))
````

Напишите рекурсивную функцию is_palindrome(s), которая проверяет, является ли строка s палиндромом (читается одинаково слева направо и справа налево).

````python
def is_palindrome(s):
    if len(s)<=1:
        return True
    return s[0]==s[-1] and is_palindrome(s[1:-1])

print(is_palindrome('ababa'))
````

Генерация всех подмножеств print(subsets([1, 2]))  # [[], [2], [1], [1, 2]]

````python
def subsets(nums):
    if not nums:
        return [[]]
    rest = subsets(nums[1:])
    return rest + [[nums[0]] + subset for subset in rest]

nums = [1,2]

subsets(nums)  # [[], [2], [1], [1, 2]]
````

1. ***Четные цифры числа***  
Напишите функцию, которая рекурсивно вычисляет количество четных цифр числа. Сделайте второй вариант этой функции – через хвостовую рекурсию.


````python
def count_dig(n):
    n=abs(n)
    if n<10:
        return 1 if n%2==0 else 0
    return (1 if (n%10)%2==0 else 0) + count_dig(n//10)

print(count_dig(-1234))
````

````python
from datetime import datetime

def transform_to_dict(s: str):
    split_info = s.split()
    if len(split_info)!=4:
        raise ValueError('Неверное к-во параметров')
    l_name, f_name, birth, course = split_info
    if not l_name.isalpha():
        raise ValueError('Только буквы')
    if not f_name.isalpha():
        raise ValueError('Только буквы')
    if not birth.isdigit():
        raise ValueError('Только целое число')
    if not 1990<=int(birth)<=datetime.now().year - 10:
        raise ValueError('Только диапазон 1990 - 2026 минус 10 лет')
    if not course.isdigit():
        raise ValueError('Только целое число')
    if not 1<=int(course)<=6:
        raise ValueError('Только диапазон 1-6')
    
    d = {}
    d['last_name'] = l_name
    d['first_name'] = f_name
    d['birth'] = birth
    d['course'] = course
    return d

transform_to_dict('Smith John 2000 3')
````

3. ***Сумма чисел списка***  
Напишите рекурсивную функцию, которая вычисляет сумму всех чисел в списке.  
Функция должна проверять:  
Аргумент должен быть списком.  
Все элементы списка должны быть числами.  
Если данные не валидны необходимо выбрасывать исключение. При вызове функции обработайте возможное исключение.

*Данные:*   
`numbers = [1, 2, 3, 4, 5]`  
Пример вывода:   
15


````python
def rec_sum(lst: list):
    if not isinstance(lst, list):
        raise TypeError('Ожидаю список чисел')
    if not isinstance(lst[0], (int, float)):
        raise TypeError('Ожидаю элементы списка как числа')
    if not lst:
        return 0
    return lst[0] + rec_sum(lst[1:])

numbers = [1, 2, 'dgh', 3, 4, 5]
try:
    print(rec_sum(numbers))
except (TypeError, ValueError) as e:
    print('Ошибка:' , e)
````
