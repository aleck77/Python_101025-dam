2. **Фильтрация по количеству**  
Напишите программу, которая фильтрует список кортежей, оставляя только те, где количество больше 10. Выведите данные согласно примеру.  
Данные:  
`data = [("апельсин", 5), ("вишня", 12), ("киви", 18), ("ананас", 7), ("яблоко", 20)]`
```
Пример вывода:
Отфильтрованные данные:
Товар : вишня     |   Цена : 12
Товар : киви      |   Цена : 18

```

````python
data = [("апельсин", 5), ("вишня", 12), ("киви", 18), ("ананас", 7), ("яблоко", 200)]
print("Отфильтрованные данные:")
for name, price in data:
    if price > 10:
        print(f"Товар : {name:<10} |   Цена : {price:>5}")
````

3. **Товары со скидкой**  
Напишите программу, которая выводит таблицу с названием товара, его ценой, размером скидки (запрашивается у пользователя), и итоговой ценой с округлением до двух знаков после запятой. Таблица должна соответствовать форматированию ниже.  
Данные:  
```
products = [
    ("Кофе", 127.99),
    ("Чай", 52.49),
    ("Шоколад", 81.99)
]
```
Пример вывода:  
Введите процент скидки: 17  

```
Название        |       Цена |     Скидка |   Итоговая цена
------------------------------------------------------------
Кофе            |     127.99 |        17% |          106.23
Чай             |      52.49 |        17% |           43.57
Шоколад         |      81.99 |        17% |           68.05
```

````python
products = [
    ("Кофе", 127.99),
    ("Чай", 52.49),
    ("Шоколад", 81.99)
]
discont = int(input('Введите процент скидки:'))

print('Название        |       Цена |     Скидка |   Итоговая цена')
print('-'*(len('Название        |       Цена |     Скидка |   Итоговая цена')+1))

for name, price in products:
    final_price = price * (1 - discont / 100)
    print(f'{name:<15} | {price:>10.2f} | {discont:>9}% | {final_price:>15.2f}')
````

4. **Таблица покупок**  
Напишите программу, которая обрабатывает список кортежей, где каждый кортеж содержит информацию о товаре: название, цена и количество.  
Выведите данные в формате таблицы, выровняв название по левому краю, а цену и количество — по правому.  
Учитывайте, что названия могут быть длинными. Выстраивайте выравнивание таблицы динамически, исходя из самого длинного названия товара.

Данные:  
```
products = [
    ("Молоко", 3, 10),
    ("Хлеб", 1, 5),
    ("Сахар-песок", 2, 15),
    ("Подсолнечное масло", 5, 3),
    ("Шоколадный батончик", 4, 20),
    ("Мороженое", 3, 20)
]
```
Пример вывода:  

```
| Название               |    Цена | Количество |
-------------------------------------------------
| Молоко                 |       3 |         10 |
| Хлеб                   |       1 |          5 |
| Сахар-песок            |       2 |         15 |
| Подсолнечное масло     |       5 |          3 |
| Шоколадный батончик    |       4 |         20 |
| Мороженое              |       3 |         20 |
```

````python
products = [
    ("Молоко", 3, 10),
    ("Хлеб", 1, 5),
    ("Сахар-песок", 2, 15),
    ("Подсолнечное масло", 5, 3),
    ("Шоколадный батончик", 4, 20),
    ("Мороженое", 3, 20)
]

max_len = 0
for item in products:
    name_len = len(item[0])
    if name_len > max_len:
        max_len = name_len

# расчет длин выравнивания
f_name_len = max_len + 3
f_price_len = 7
f_quantity_len = 12
line_len = f_name_len + f_price_len + f_quantity_len + 10

#заголовок
print(f'| {'Название':<{f_name_len}} | {'Цена':>{f_price_len}} | {'Количество':>{f_quantity_len}} |')
print('-'*line_len)

#вывод табл. данных
for name, price, quantity in products:
    print(f'| {name:<{f_name_len}} | {price:>{f_price_len}} | {quantity:>{f_quantity_len}} |')
````

```

```

````python
clients = [
    ("Иван", "Иванов", 10, [("Кофе", 100.00), ("Чай", 200.00)]),
    ("Анна", "Смирнова", 15, [("Шоколад", 150.00), ("Печенье", 50.00)]),
    ("Петр", "Сидоров", 20, [("Чайник", 1000.00), ("Кружка", 300.00)])
]

for name, last_name, discount, lst_purchases in clients:
    print(f'Уважаемый(ая) {name} {last_name},')
    print()
    print('Спасибо за покупку в нашем магазине!')
    print()
    print('Вы купили следующие товары:')
    sum_purch = 0
    for name_prod, price in lst_purchases:
        sum_purch += price
        print(f'\t- {name_prod}: {price}')
    print()
    print(f'Общая сумма покупок: {sum_purch}')
    print(f'Ваша скидка: {discount}%')
    print(f'Сумма с учётом скидки: {sum_purch * (1 - discont / 100):.2f}')
    print()
    print('Благодарим за покупку!')
    print('_'*60)

````

```
Особенности вывода:
```

`participants = [

````python
participants = [
    ("Иван", "Иванов", "Бег на 100м", 1, (0, 9.58)),
    ("Мария", "Петрова", "Плавание", 2, (1, 2.34)),
    ("Сергей", "Сидоров", "Велогонка", 3, (25, 15.10)),
    ("Анна", "Смирнова", "Марафон", 1, (120, 34.56)),
    ("Дмитрий", "Ковалёв", "Прыжки в длину", 4, (0, 8.25))
]

compliments = [
    "Удачи в следующий раз!",          
    "Поздравляем с победой!",           # 1 место
    "Отличный результат!",              # 2 место
    "Вы заняли призовое место!"         # 3 место
]

for name, surname, sport, place, (minute, sec) in participants:
    if place in (1,2,3):
        message = compliments[place]
    else:
        message = compliments[0]      
    print(f'{name} {surname}, спасибо за участие в {sport}! {message}')
````

````python
participants = [
    ("Иван", "Иванов", "Бег на 100м", 1, (0, 9.58)),
    ("Мария", "Петрова", "Плавание", 2, (1, 2.34)),
    ("Сергей", "Сидоров", "Велогонка", 3, (25, 15.10)),
    ("Анна", "Смирнова", "Марафон", 1, (120, 34.56)),
    ("Дмитрий", "Ковалёв", "Прыжки в длину", 4, (0, 8.25))
]
#поздравления
for name, surname, sport, place, (minute, sec) in participants:
    if place == 1:
        message = 'Поздравляем с победой!'
    elif place == 2:
        message = 'Отличный результат!'
    elif place == 3:
        message = 'Вы заняли призовое место!'
    else:
        message = 'Удачи в следующий раз!'        
    print(f'{name} {surname}, спасибо за участие в {sport}! {message}')
print()
#таблица
print('ОТЧЁТ ПО МЕРОПРИЯТИЮ:')
print(f'{'Имя':<10}{'Фамилия':<15}{'Вид спорта':<20}{'Место':<10}{'Время завершения':>16}')
print('-'*71)

total_sec = 0
for name, surname, sport, place, (minute, sec) in participants:
    print(f'{name:<10}{surname:<15}{sport:<20}{place:<10}{minute:>10}:{sec:05.2f}')
    total_sec += minute * 60 + sec
print('-'*71)
print(f'Общее количество участников: {len(participants)}')
avg_sec = total_sec / len(participants)
print(f'Среднее время завершения: {int(avg_sec // 60) } мин {avg_sec % 60:.2f} сек')
````
